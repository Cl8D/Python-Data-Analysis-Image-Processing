```
import requests

# 특정 URL에 접속하는 request 객체 생성하기
request = requests.get('https://www.naver.com/')

# 접속한 이후의 웹 사이트 소스코드 추출하기
html = request.text.strip() #불필요한 공백 제거하기

print(html)
```

```
import requests
#beautifulSoup 활용하기
from bs4 import BeautifulSoup

request = requests.get('http://www.dowellcomputer.com/main.jsp')
html = request.text
# html 객체를 beautifulsoup 객체로 변경하기
soup = BeautifulSoup(html, 'html.parser')

# <a> 태그를 포함하는 요소 추출
# <a> 태그의 경우 일반적으로 링크 정보를 가지고 있다.
links = soup.select('td > a')

for link in links:
  # 링크가 href 속성을 가지고 있다면,
  if link.has_attr('href'):
    # href 속성의 값으로 notice라는 문자가 포함되어 있다면 (공지사항이라고 인식)
    if link.get('href').find('notice') != -1:
      # 공지사항 제목만 추출한다고 생각할 수 있다.
      print(link.text)
```
