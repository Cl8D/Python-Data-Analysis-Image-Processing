```
import matplotlib.pyplot as plt
import numpy as np
# 하나의 이미지 파일에 여러 개의 그래프를 넣을 수 있다.(subplot)

x = np.linspace(0, np.pi * 10, 500) # PI * 10 너비에, 500개의 점 찍기
fig, axes = plt.subplots(2, 1) # 2개의 그래프가 들어가는 Figure 생성

axes[0].plot(x, np.sin(x)) 
axes[1].plot(x, np.cos(x))

fig.savefig("sin&cos.png")
```

```
import matplotlib.pyplot as plt
import numpy as np
# 하나의 이미지 파일에 여러 개의 그래프를 넣을 수 있다.(subplot)

x = np.linspace(0, np.pi * 10, 500) # PI * 10 너비에, 500개의 점 찍기
fig, axes = plt.subplots(1, 2) # 2개의 그래프가 들어가는 Figure 생성

axes[0].plot(x, np.sin(x)) 
axes[1].plot(x, np.cos(x))

fig.savefig("sin&cos.png")
```

```
import matplotlib.pyplot as plt
import numpy as np

# x, y값을 지정
x = np.arange(-9, 10)
y = x ** 2

# 라인 스타일로는 '-', ':', '-.', '--' 등 사용 가능
plt.plot(x, y, linestyle=":", marker="*")
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y1 = x ** 2
y2 = -x

plt.plot(x, y1, linestyle="-.", marker="*", color="red", label="y = x * x")
plt.plot(x, y2, linestyle=":", marker="o", color="blue", label="y = -x")

plt.xlabel("X")
plt.ylabel("Y")

# 범례 설정
plt.legend(
  shadow=True,
  borderpad=1
)
plt.show()
```
